```python
from bs4 import BeautifulSoup
import requests
import json
import pandas as pd
import time
import datetime as dt

df = pd.DataFrame(columns=['id', 'photo', 'name', 'blurb', 'goal', 'pledged', 'state', 'slug',
       'disable_communication', 'country', 'country_displayable_name',
       'currency', 'currency_symbol', 'currency_trailing_code', 'deadline',
       'state_changed_at', 'created_at', 'launched_at', 'staff_pick',
       'is_starrable', 'backers_count', 'static_usd_rate', 'usd_pledged',
       'converted_pledged_amount', 'fx_rate', 'current_currency', 'usd_type',
       'creator', 'location', 'category', 'profile', 'spotlight', 'urls',
       'percent_funded', 'is_liked', 'is_disliked'])

page = 1
while page != 201:
    url = f'https://www.kickstarter.com/discover/advanced?sort=magic&seed=2684100&page={page}'
    r = requests.get(url)
    if r.ok == True:
        print(url)
        html_text = r.text
        tree = BeautifulSoup(html_text,'html.parser')
        projects_info = tree.find_all('div', {'class': 'js-react-proj-card grid-col-12 grid-col-6-sm grid-col-4-lg'})
        for project_info in projects_info:
            project = json.loads(project_info['data-project'])
            df = df.append(project, ignore_index=True)
        page+=1
    else:
        print("error")
        time.sleep(1)

print(len(df))
```

```python
df.to_csv("data.csv",index=False)
```
